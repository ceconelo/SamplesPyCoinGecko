```python
market_info_coin = []
for i in range(1, 103):
  data = cg.get_coins_markets(vs_currency='usd', per_page='100', page=i)
  market_info_coin.append(data)
  print('Page {}'.format(i))
```

```python
df = pd.DataFrame()
for i in range(0,102):
  df = df.append(market_info_coin[i])
  print(i)
  
```

# Utils


```python
def chuncks(lst, n):
  for i in range(0, len(lst), n):
    yield lst[i:i+n]
```

```python
def color_negative_red(value):
  if value < 0:
    color = 'red'
  elif value > 0:
    color = 'green'
  else:
    color = 'black'

  return 'color: %s' % color
```
